```python
!pip install datasets
from datasets import load_dataset
import os

# Download the IMDB dataset
dataset = load_dataset('imdb')

# Create the target directory
os.makedirs('data/imdb', exist_ok=True)

# Save in MixText format: <label>\t<text>
def write_txt(split, path):
    with open(path, 'w', encoding='utf-8') as f:
        for item in dataset[split]:
            label = item['label']
            text = item['text'].strip().replace('\n', ' ').replace('\t', ' ')
            f.write(f"{label}\t{text}\n")

write_txt('train', 'data/imdb/train.txt')
write_txt('test', 'data/imdb/test.txt')
```

```python
# Step 1: Install & Load Models en- de- en
from transformers import MarianMTModel, MarianTokenizer

# EN → DE
en2de_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-de")
en2de_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-de")
en2de_model = en2de_model.to("cuda")  # ✅ move to GPU

# DE → EN
de2en_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-de-en")
de2en_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-de-en")
de2en_model = de2en_model.to("cuda")  # ✅ move to GPU
```

```python
# en- ru- en
from transformers import MarianMTModel, MarianTokenizer


# English → Russian
en2ru_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
en2ru_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-ru").to("cuda")

# Russian → English
ru2en_tokenizer = MarianTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ru-en")
ru2en_model = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-ru-en").to("cuda")
```

```python
assert torch.cuda.is_available(), "GPU not enabled — go to Runtime → Change Runtime Type → GPU"
```

```python
# STEP2 : Back-Translation Helper Functions
import torch

def translate(texts, tokenizer, model):
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True).to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs)
    return tokenizer.batch_decode(outputs, skip_special_tokens=True)

def back_translate(texts):
    de = translate(texts, en2de_tokenizer, en2de_model)
    en = translate(de, de2en_tokenizer, de2en_model)
    return en
```

```python
def back_translate_ru(texts):
    ru = translate(texts, en2ru_tokenizer, en2ru_model)
    en = translate(ru, ru2en_tokenizer, ru2en_model)
    return en
```

```python
#  Step 3: Load IMDB sentences and translate
import pandas as pd

# Load from your earlier IMDB `train.txt` file (only unlabeled part if you want)
df = pd.read_csv("data/imdb/train.txt", sep='\t', header=None)
texts = df[1].tolist()

# Sample subset for now (can expand later)
subset = texts[:1000]  # or 25000 if you want full

# Batch translate and collect
bt_results = []
batch_size = 16

for i in range(0, len(subset), batch_size):
    batch = subset[i:i+batch_size]
    bt_batch = back_translate(batch)
    bt_results.extend(bt_batch)

print("Translated:", len(bt_results))
```

```python
import pickle

with open("data/imdb/de_1.pkl", "wb") as f:
    pickle.dump(bt_results, f)

print("✅ Saved de_1.pkl!")
```

```python
import pandas as pd

df = pd.read_csv("data/imdb/train.txt", sep='\t', header=None)
texts = df[1].tolist()

subset = texts[:1000]  # 你可以从小规模开始试跑
bt_results_ru = []

batch_size = 16
for i in range(0, len(subset), batch_size):
    batch = subset[i:i+batch_size]
    bt_batch = back_translate_ru(batch)
    bt_results_ru.extend(bt_batch)

print("Translated (EN→RU→EN):", len(bt_results_ru))
```

```python
import pickle

with open("data/imdb/ru_1.pkl", "wb") as f:
    pickle.dump(bt_results_ru, f)

print("✅ Saved ru_1.pkl!")
```
